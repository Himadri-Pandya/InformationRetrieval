```python
'''
The following code takes an input file containing multiple documents represented within <TEXT> </TEXT> tags
It extracts words from each documents, counts word frequency within each document and creates a word_to_vector matrix
word_to_vector has no.of rows = no. of text fields in input file and no.of columns = no. words in word space
Finally it writes this matrix to an output file.
'''

import xml.etree.ElementTree as ET
import itertools
import lxml.etree as etree
import re
from collections import Counter

parser = etree.XMLParser(recover=True)
word_vector = []
dimension = []
word_freq_in_doc = []
word_to_vector = []

#Open the input file, enclose the contents between a fake <root> </root> tag
with open("/home/muskan_2117/DAIICT/Sem2/IT550/InformationRetrieval/word_to_vector_input_wsj7_001") as file:
        with_fake_root = itertools.chain('<root>', file, '</root>')
        fake_root = ET.fromstringlist(with_fake_root,parser=parser)
        
#Get the word_vector and dimenstion of word space from text feild enclosed between every <DOC> tags         
for doc in fake_root.findall('DOC'):
    temp = []
    doc_vector = []
    for word in doc.iter():
        temp.append(word.text)
        for i in temp:
            str = i
            #Extract words from text using regular expression
            if i != '\n' and str.startswith("\n"):
                i = re.compile('\w+').findall(i)
                doc_vector.extend(i) 
    word_vector.append(doc_vector)
    dimension.extend(doc_vector)
    
#Dimension should have every word apeared in text fields exactly once 
dimension = list(set(dimension))

#Get all zero row vector of size len(dimension)
zeros = []
for i in range(len(dimension)):
    zeros.append(0)
    
temp = zeros.copy()

#Count word frequency and create word_to_vector matrix,
#word_to_vector has no.of rows = no. of text fields in input file and no.of columns = no. words in word space

for vector in word_vector:
    word_freq_in_doc = Counter(vector)
    for word in vector:
        word_count = word_freq_in_doc[word]
        i = dimension.index(word)
        temp[i] = word_count
    word_to_vector.append(temp)
    temp = zeros.copy()

#Write the word_to_vector matrix to output file    
with open('/home/muskan_2117/DAIICT/Sem2/IT550/InformationRetrieval/word_to_vector_output.txt', 'w') as f:
    for item in word_to_vector:
        f.write("%s\n" % item)


```
